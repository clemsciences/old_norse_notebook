```python
import numpy as np
import scipy

import skimage.io as skio
import skimage.draw as skdr
import skimage.feature as skf
import skimage.filters as skfi
import skimage.transform as sktr
import skimage.morphology as skmp
from skimage import img_as_uint, img_as_int
from matplotlib import pyplot as plt


import warnings
warnings.simplefilter("ignore")

%matplotlib inline
```

First page of the first grammatical treatise in the Codex Wormianus

```python
image = skio.imread("siden1denfoerstegrammatiskeavhandlingen.png", as_grey=True)
skio.imshow(image)
skio.show()
```

```python
skio.imshow(image[20:75, 100:250])
skio.show()
```

```python
ca_image = np.uint8(skf.canny(image))*255
skio.imshow(ca_image[20:75, 100:250])
skio.show()
```

```python
ca_image = np.uint8(skf.canny(image))*255
thresh = skfi.threshold_otsu(ca_image)
th_image = ca_image > thresh
sk_image = skmp.skeletonize(th_image[20:75, 100:250])
skio.imshow(sk_image)
skio.show()
```

```python
thresh = skfi.threshold_otsu(image)
th_image = image > thresh
skio.imshow(th_image[20:75, 100:250])
skio.show()
```

```python
thresh = skfi.threshold_mean(image)
th_image = image > thresh
skio.imshow(th_image[20:75, 100:250])
skio.show()
```

```python
fig, ax = skfi.try_all_threshold(image, figsize=(10, 8), verbose=False)
plt.show()
```
